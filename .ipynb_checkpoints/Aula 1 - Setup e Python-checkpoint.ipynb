## venv: Ambientes virtuais

* Criação de ambiente virtual

`python -m venv venv`

* Ativação/desativação de ambiente virtual

Ativação: `source ./venv/bin/activate`

Desativação: `deactivate`

Obs: VSCode detecta automatico
