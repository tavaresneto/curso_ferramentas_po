# Trabalhando com arquivos texto

```python
# Classe que vai receber os dados
class Problem_LS:
    def __init__(self):
        self.setup_cost = 0
        self.holding_cost = 0
        self.demand = []

    def __str__(self):
        ret = f"""Setup cost: {self.setup_cost}
Holding cost: {self.holding_cost}
i\tdi\n"""
        for i,d in enumerate(self.demand):
            ret = ret + f"{i}\t{d}\n"
        return ret

    def load_from_file(self, filename):
        with open(filename, "r") as fobj:
            linha = fobj.readline().split()
            print("A primeira linha do arquivo já dividida é:", linha)
            self.setup_cost = float(linha[1])

            linha = fobj.readline().split()
            self.holding_cost = float(linha[1])

            linha = fobj.readline().split() #pula linha 3 do arquivo
            for linha in fobj:
                self.demand.append(float(linha.split()[1]))
```

```python
problem = Problem_LS()
problem.load_from_file("arquivos_aula_4/dados_lot_sizing.txt")
print(problem)
```

```python
import pandas as pd
#Se você não tem o Pandas instalado => python3 -m pip install pandas Jinja2 openpyxl

# Dados para o DataFrame
lista_cidade = ['New York, USA', 'Tokyo, Japan', 'Paris, France', 'Sydney, Australia', 
 'Rio de Janeiro, Brazil', 'Cairo, Egypt', 'Berlin, Germany', 'Moscow, Russia', 
 'Beijing, China', 'Cape Town, South Africa', 'Toronto, Canada', 'Buenos Aires, Argentina', 
 'Mexico City, Mexico', 'Istanbul, Turkey', 'Bangkok, Thailand']
lista_latitude = [40.7128, 35.6762, 48.8566, -33.8688, -22.9068, 30.0444, 52.52, 55.7558, 
 39.9042, -33.9249, 43.6511, -34.6037, 19.4326, 41.0082, 13.7563]
lista_longitude = [-74.006, 139.6503, 2.3522, 151.2093, -43.1729, 31.2357, 13.405, 37.6173, 
 116.4074, 18.4241, -79.3837, -58.3816, -99.1332, 28.9784, 100.5018]
lista_vontade = [1, 7, 8, 5, 7, 0, 2, 5, 0, 10, 5, 10, 9, 9, 10]

# Criação do DataFrame
df = pd.DataFrame(
    {
        "Cidade": lista_cidade,
        "Latitude": lista_latitude,
        "Longitude": lista_longitude,
        "Vontade de visitar": lista_vontade
    }
)

df.head()
```

# Juntando tabelas

```python
# Lista de países
country_list = ['USA', 'Japan', 'France', 'Australia', 'Brazil', 'Egypt', 'Germany', 
                'Russia', 'China', 'South Africa', 'Canada', 'Argentina', 'Mexico', 
                'Turkey', 'Thailand']

# Correspondente DDI (código de chamada internacional) para cada país
call_codes = ['+1', '+81', '+33', '+61', '+55', '+20', '+49', 
              '+7', '+86', '+27', '+1', '+54', '+52', '+90', '+66']

# Criando o DataFrame
df_ddi = pd.DataFrame({
    'Country': country_list,
    'Call Code': call_codes
})
```

Queremos juntar df e df_ddi para ter uma tabela com "Call Code" e "Cidade":

#

```python
df_2 = pd.merge(df, df_ddi, how="inner", left_on="País", right_on="Country")

print(df_2.head())

df_2[["Cidade", "Call Code"]]
```
